```python
#Q - 3. What is a box tuple, and how does it work?
#Answer:
In the context of image processing libraries like Pillow, a box tuple is a tuple that represents a rectangular region or bounding box. It is commonly used to specify the coordinates and size of a rectangular area within an image.

A box tuple typically consists of four integer values: (left, upper, right, lower). Here's an explanation of each value:

left: The x-coordinate of the left edge of the box.
upper: The y-coordinate of the upper edge of the box.
right: The x-coordinate of the right edge of the box.
lower: The y-coordinate of the lower edge of the box.
```

```python
#Q - 4. Use your image and load in notebook then, How can you find out the width and height of an Image object?
#Answer:
#To find out width and height of image object using pillow library we have to use 'Image.size' attribute
from PIL import Image
import matplotlib.pyplot as plt

image = Image.open("/Users/Moti/Desktop/indian girl wit 0 (1).png") #Opening the image

#displaying the image
plt.imshow(image)
plt.axis('off')
plt.show()

#Get the width and height of an image
width , height = image.size

#print width and height
print("width:", width)
print("height: ", height)
```

```python
#Q - 8. Image objects do not have drawing methods. What kind of object does? How do you get this kind of object?
#Answer:
Image objects in the Pillow library do not have built-in drawing methods. Instead, the drawing methods are available through the ImageDraw object, which is obtained by calling the ImageDraw.Draw() function with an Image object as a parameter.
```
